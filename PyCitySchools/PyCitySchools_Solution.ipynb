```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(100)
```

## District Summary

```python
student_data.head()
```

```python
# Calculate the total number of unique schools


school_count = school_data['school_name'].nunique()
school_count


print(f"Total Number of Unique School : {school_count}") 
```

```python
# Calculate the total number of students
student_count = student_data['Student ID'].nunique()
student_count

print(f"Total students : {student_count}")
```

```python
# Calculate the total budget                               
total_budget =school_data['budget'].unique().sum()
total_budget
print(f"Total budget : {total_budget}")
```

```python
# Calculate the average (mean) math score
average_math_score =student_data['math_score'].mean()
average_math_score
print(f"Average math score :  {average_math_score} " )
```

```python
# Calculate the average (mean) reading score
average_reading_score =student_data['reading_score'].mean()
average_reading_score

print(f"Average reading score :  {average_reading_score} " )
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = student_data[(student_data["math_score"] >= 70)].count()["student_name"]

passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage


print(f"% passing math (the percentage of students who passed math) : {passing_math_percentage}")
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)
passing_reading_count = student_data[(student_data["reading_score"] >= 70)].count()["student_name"]

passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage



print(f"% passing reading (the percentage of students who passed reading ) : {passing_reading_percentage}")
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = student_data[
    (student_data["math_score"] >= 70) & (student_data["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate


print(f"% overall passing (the percentage of students who passed math AND reading) : {overall_passing_rate}" )

```

```python
import pandas as pd


# Create a high-level snapshot of the district's key metrics in a DataFrame
data = { 
    "Total Students": student_count,
    "Total Budget": total_budget,
     "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "passing_math_percentage": passing_math_percentage,
    "passing_reading_percentage": passing_reading_percentage,
    "overall_passing_rate": [overall_passing_rate]
    }

#convert a dictionary into adataframe
district_summary = pd.DataFrame(data)

#format the columns
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("{:,.2f}".format)
district_summary["Average Math Score"] = district_summary ["Average Math Score"].map("{:,.2f}".format)
district_summary["Average Reading Score"] = district_summary ["Average Reading Score"].map("{:,.2f}".format)
district_summary["passing_math_percentage"] = district_summary ["passing_math_percentage"].map("{:,.2f}%".format)
district_summary["passing_reading_percentage"]=district_summary["passing_reading_percentage"].map("{:,.2f}%".format)
district_summary["overall_passing_rate"]= district_summary["overall_passing_rate"].map("{:,.2f}%".format)




# Display the DataFrame
district_summary
```
